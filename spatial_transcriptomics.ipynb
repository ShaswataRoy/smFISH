```python
gapdh_reinduction_mean = np.array([np.mean(x.groupby(["tile","cell"]).size().to_numpy()) for x in gapdh_reinduction])
gapdh_reinduction_std = np.array([np.std(x.groupby(["tile","cell"]).size().to_numpy()) for x in gapdh_reinduction])

plt.xlabel("Mean")
plt.ylabel("Standard deviation")
plt.plot(np.arange(0,15000,100),np.arange(0,15000,100))
```

```python
gbp5_reinduction_mean = np.array([np.mean(x.groupby(["tile","cell"]).size().to_numpy()) for x in gpb5_reinduction])
gbp5_reinduction_std = np.array([np.std(x.groupby(["tile","cell"]).size().to_numpy()) for x in gpb5_reinduction])
plt.plot(gbp5_reinduction_mean, gbp5_reinduction_std,'ro',label="GBP5 Reinduction")
plt.plot(gapdh_reinduction_mean, gapdh_reinduction_std,'go',label="GAPDH Reinduction")
plt.xlabel("Mean")
plt.ylabel("Standard deviation")
plt.plot(np.arange(0,8000,100),np.arange(0,8000,100))
plt.xlim(0,25000)
plt.ylim(0,25000)
plt.legend()
```
